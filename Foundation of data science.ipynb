```python
import pandas as pd

data = pd.read_csv('happyscore_income.csv')
```

```python
reg_gdp = data.groupby("region")['GDP'].mean()
reg_happyscore = data.groupby("region")['happyScore'].mean()
```

```python
region_df = pd.merge(left=reg_gdp, right=reg_happyscore, on="region", how="left", left_index=False, right_index=False,)
region_df = region_df.reset_index()
region_df.head()
```

```python
mean_gdp_region = region_df['GDP']
mean_happy_region = region_df['happyScore']
```
