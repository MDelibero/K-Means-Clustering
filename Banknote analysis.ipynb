```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('banknote_dataset')
```

```python
data.describe()
```

```python
V1 = data['V1']
V2 = data['V2']

plt.xlabel('V2')
plt.ylabel('V1')
plt.scatter(V1, V2, alpha = 0.7)

plt.savefig("scatter_plot.png")
```

```python
from sklearn.cluster import KMeans
import numpy as np
V1_V2 = np.column_stack((V1, V2))

km_res = KMeans(n_clusters=3).fit(V1_V2)

clusters = km_res.cluster_centers_
plt.scatter(V1,V2)
plt.scatter(clusters[:,0], clusters[:,1], s=1000)
plt.savefig("K-Means Clustering.png")
```
